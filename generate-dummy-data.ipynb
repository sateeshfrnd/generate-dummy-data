```python
import pandas as pd
from faker import Faker
```

```python
# Create a Faker object
fake = Faker()
```

```python
def generate_data(num_records):
    return pd.DataFrame(
        {
            'Name': [fake.name() for _ in range(num_records)],
            'Email': [fake.email() for _ in range(num_records)],
            'Address': [fake.address() for _ in range(num_records)],
            'PhoneNumber': [fake.phone_number() for _ in range(num_records)],
            'Country': [fake.country() for _ in range(num_records)],
            'Latitude': [fake.latitude() for _ in range(num_records)],
            'Longitude': [fake.longitude() for _ in range(num_records)],
            'DateofBirth': [fake.date_of_birth(minimum_age=18, maximum_age=90) for _ in range(num_records)],
            'Company': [fake.company() for _ in range(num_records)],
            'Job': [fake.job() for _ in range(num_records)],
            'CreditCardNumber': [fake.credit_card_number() for _ in range(num_records)],
            'ExpirationDate': [fake.credit_card_expire() for _ in range(num_records)],
        }
    )  
```

```python
generate_data(10)
```

```python
generate_data(20)
```

```python
# Save DataFrame to a CSV file
generate_data(100).to_csv('fake_data.csv', index=False)
```

```python
# Save DataFrame to an Excel file
generate_data(200).to_excel('fake_data.xlsx', index=False)
```

```python
# Save DataFrame to a JSON file
generate_data(10).to_json('fake_data.json', orient='records')
```
